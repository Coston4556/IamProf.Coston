```python
import os
import cv2
import shutil

# Define root folders [ Change the names depending on your current structure ]
# The image will be resized to height of 64 and width of 64

dataset = "./Fruit"
processed_dataset = "./ProcessedFruit"
resize_width = 64 
resize_height = 64

# Clear output folder
if os.path.exists(processed_dataset):
    shutil.rmtree(processed_dataset)
os.makedirs(processed_dataset, exist_ok=True)

# Walk through input folder recursively
for root, _, files in os.walk(dataset):
    for filename in files:
        if filename.lower().endswith((".jpg", ".jpeg", ".png")):
            input_path = os.path.join(root, filename)

            # Construct output path preserving subfolders
            relative_path = os.path.relpath(input_path, dataset)
            output_path = os.path.join(processed_dataset, relative_path)
            output_dir = os.path.dirname(output_path)
            os.makedirs(output_dir, exist_ok=True)

            # Load image from folder
            image = cv2.imread(input_path)
            if image is None:
                print(f"Skipped unreadable file: {input_path}")
                continue

            # Resize image
            image_resized = cv2.resize(image, (resize_width, resize_height))

            # Denoise (only if colored, otherwise use cv2.fastNlMeansDenoising)
            final_image = cv2.fastNlMeansDenoisingColored(image_resized, None, 10, 10, 7, 21)

            # Grayscale, uncomment the following line if you choose to grayscale your images
            # final_image = cv2.cvtColor(final_image, cv2.COLOR_BGR2GRAY)

            # Save grayscale image
            cv2.imwrite(output_path, final_image)
```
