```python
# This is how you import functions from your packages

from cv2 import imread, cvtColor, fastNlMeansDenoisingColored, resize, COLOR_BGR2GRAY, COLOR_BGR2RGB
from matplotlib import pyplot

# if you add new imports, you will have to replay this block.
```

```python
original_image = imread("./myimage.png") # Opens an image in Blue, Green, Red Mode (BGR)
rgb_image = cvtColor(original_image, COLOR_BGR2RGB) # Changes the color space to RGB ( Red, Green Blue )

# Set up a 2 column image graph
pyplot.subplot(1,2,1)
pyplot.imshow(original_image)

pyplot.subplot(1,2,2)
pyplot.imshow(rgb_image)

# Show the graph
pyplot.show()
```

```python
# This is an example of grayscaling an image

grayscale_image = cvtColor(rgb_image, COLOR_BGR2GRAY)
pyplot.subplot(1,2,1)
pyplot.imshow(rgb_image)

pyplot.subplot(1,2,2)
pyplot.imshow(grayscale_image, cmap="gray")
pyplot.show()
```

```python
# This is an example of denoising an image

noisy_image = imread("./noisy.jpeg")
noisy_image_rgb = cvtColor(noisy_image, COLOR_BGR2RGB) # Change the color space to RGB ( Red, Green Blue )

pyplot.subplot(1,2,1)
pyplot.imshow(noisy_image_rgb)

denoised_image = fastNlMeansDenoisingColored(noisy_image_rgb, None, 10, 10, 7, 100)

pyplot.subplot(1,2,2)
pyplot.imshow(denoised_image)
pyplot.show()
```

```python
# This is an example of resizing an image

resized_image = resize(rgb_image, (224,224))

pyplot.subplot(1,2,1)
pyplot.imshow(rgb_image)

pyplot.subplot(1,2,2)
pyplot.imshow(resized_image)
pyplot.show()
```

```python
import albumentations as A
```

```python
transform = A.Compose([
    A.HorizontalFlip(p=1),
    A.RandomBrightnessContrast(p=0.4),
    A.Rotate(limit=20, p=0.5),
    A.RandomCrop(width=64, height=64, p=0.3),
])

augmented = transform(image=rgb_image)["image"]
```

```python
pyplot.subplot(1,2,1)
pyplot.imshow(rgb_image)

pyplot.subplot(1,2,2)
pyplot.imshow(augmented)
pyplot.show()
```
